```python
main_table  =  soup.find_all('table', class_='MsoNormalTable')[2]
```

```python
column_titles = main_table.find_all('th')
```

```python
def clean(text):
    text = text.replace('\n', '')
    text = text.replace('\t', '')
    text = text.replace(' ', '')
    text = text.strip()
    return text
```

```python
def clean_data(text):
    text = text.replace('\n', '')
    text = text.replace('\t', '')
    text = text.strip()
    return text
```

```python
titles = [clean(title.text) for title in column_titles]
```

```python
titles
```

```python
import pandas as pd
df = pd.DataFrame(columns=titles)
```

```python
df.head()
```

```python
rows = main_table.find_all('tr')[1:]
rows
```

```python
for row in rows:
    datas = row.find_all('td')
    row_data = [clean_data(data.text) for data in datas]
    df.loc[len(df)] = row_data
```

```python
def scrape_data(link):
    html = requests.get(link, verify=False)
    soup = BeautifulSoup(html.text, 'html.parser')
    main_table = soup.find_all('table', class_='MsoNormalTable')[2]
    header = soup.find_all('table', class_='MsoNormalTable')[1].find('strong').text
    rows = main_table.find_all('tr')[1:]
    error = False
    for row in rows:
        datas = row.find_all('td')
        row_data = [clean_data(data.text) for data in datas]
        new_row = pd.Series(row_data, index=df.columns)
        df.loc[len(df)] = row_data
```

```python
def check_data(link):
    html = requests.get(link, verify=False)
    soup = BeautifulSoup(html.text, 'html.parser')
    main_table = soup.find_all('table', class_='MsoNormalTable')[2]
    header = soup.find_all('table', class_='MsoNormalTable')[1].find('strong').text
    rows = main_table.find_all('tr')[1:]
    error = False
    for row in rows:
        datas = row.find_all('td')
        row_data = [clean_data(data.text) for data in datas]
        #new_row = pd.Series(row_data, index=df.columns)
        #df.loc[len(df)] = row_data
        if (len(row_data) != 6):
            error = True
    if error:
        print("There has been an error")
    else:
        print("There has been no error")
```
