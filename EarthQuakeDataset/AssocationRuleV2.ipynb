```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('earthquake_dataset.csv')
```

```python
df.head()
```

In order to separate each category into their date we have to combine day and months into a single column named "Day Count"

```python
df['Day Count'] = 0 #Making the column
df.head()
```

```python
monthsToDays = {
    4: 91,
    3: 60,
    2: 31,
    1: 0
} #The months will have the corresponding number of days before their actual month like in april: january (31) + february (29)
  # + march (31) = 61
monthsToDays[2]
```

```python
for index, row in df.iterrows():
    df.loc[index, 'Day Count'] = row['Day'] + monthsToDays[row['Month']] - 1 #this is for the index
df[1000:1010] #To check the day count
```

```python
dayCount_unique = df['Day Count'].unique()
print(sorted(dayCount_unique))
```

```python
association_list = []
for i in dayCount_unique:
    association_list.append([])
print(len(association_list)) #NOW WE HAVE THE LIST
```

```python
 for index, row in df.iterrows():
        association_list[row['Day Count']].append(row["Richter Category"]) 
        #This will only append to the list of the day count it is corresponding to
        
print(association_list)
```

```python
count = 0
for i in association_list:
    count += len(i)
print(count) #should equate to the length of the original dataframe
print(len(df)) #This is to check if we  have the right amount of data
```

```python
categories = ['micro','minor', 'light', 'moderate', 'strong', 'major', 'great']
association_df = pd.DataFrame(columns=categories)
association_df.head()
```

```python
for i, row in enumerate(association_list):
    row_data = []
    for category in categories:
        if (category in row):
            row_data.append(1)
        else:
            row_data.append(0)
    association_df.loc[len(association_df)] = row_data
association_df.head()
```

```python
association_df.to_csv('association_v2.csv', index=False)
df.to_csv('earthquake_dataset.csv', index=False)
```

```python
dataset = pd.read_csv('association_v2.csv')
dataset.tail()  #read the CSV file
```

```python
#find frequently occurring itemsets using Apriori Algorithm
from mlxtend.frequent_patterns import apriori
frequent_itemsets_ap = apriori(dataset, min_support=0.1, use_colnames=True)
frequent_itemsets_ap
```

```python
#find frequently occurring itemsets using FP-growth Algorithm
#same identical items and support with Apriori algorithm
from mlxtend.frequent_patterns import fpgrowth
frequent_itemsets_fp = fpgrowth(dataset, min_support=0.1, use_colnames=True)
frequent_itemsets_fp
```

```python
#mining the Association Rules using Apriori and FP-growth algorithms
from mlxtend.frequent_patterns import association_rules
rules_ap = association_rules(frequent_itemsets_ap, metric="confidence", min_threshold=0.8)
rules_ec = association_rules(frequent_itemsets_ap, metric="support", min_threshold=0.2)
rules_fp = association_rules(frequent_itemsets_fp, metric="confidence", min_threshold=0.8)
```
