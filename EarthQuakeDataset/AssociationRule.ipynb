```python
def getRichter(magnitude):
    if (magnitude < 3):
        return "micro"
    elif (magnitude < 4):
        return "minor"
    elif (magnitude < 5):
        return "light"
    elif (magnitude < 6):
        return "moderate"
    elif (magnitude < 7):
        return "strong"
    elif (magnitude < 8):
        return "major"
    else:
        return "great"
```

```python
for index, row in df.iterrows():
    df.loc[index, 'Richter Category'] = getRichter(row['Mag'])
df.head(23)
```

```python
unique_province = df['Province'].unique()
print(sorted(unique_province)) #GET THE LIST OF PROVINCES
```

```python
association_list = []
for i in unique_province:
    association_list.append([])
print(association_list) #NOW WE HAVE THE LIST
```

```python
 for index, row in df.iterrows():
        association_list[row['Province']].append(row["Richter Category"]) 
        #This will only append to the list of the province it is corresponding to
        
print(association_list)
```

```python
count = 0
for i in association_list:
    count += len(i)
print(count) #should equate to the length of the original dataframe
print(len(df)) #This is to check if we  have the right amount of data
```

```python
association_df = pd.DataFrame(columns=['micro','minor', 'light', 'moderate', 'strong', 'major', 'great'])
```

```python
categories = ['micro','minor', 'light', 'moderate', 'strong', 'major', 'great']
for i, row in enumerate(association_list):
    row_data = []
    for category in categories:
        if (category in row):
            row_data.append(1)
        else:
            row_data.append(0)
    association_df.loc[len(association_df)] = row_data
```

```python
association_df.head()
```

```python
association_df.to_csv('association.csv', index=False)
df.to_csv('earthquake_dataset.csv', index=False)
```

```python
dataset = pd.read_csv('association.csv')
dataset.tail()  #read the CSV file
```

```python
#find frequently occurring itemsets using Apriori Algorithm
from mlxtend.frequent_patterns import apriori
frequent_itemsets_ap = apriori(dataset, min_support=0.1, use_colnames=True)
frequent_itemsets_ap
```

```python
#find frequently occurring itemsets using FP-growth Algorithm
#same identical items and support with Apriori algorithm
from mlxtend.frequent_patterns import fpgrowth
frequent_itemsets_fp = fpgrowth(dataset, min_support=0.1, use_colnames=True)
frequent_itemsets_fp
```

```python
#mining the Association Rules using Apriori and FP-growth algorithms
from mlxtend.frequent_patterns import association_rules
rules_ap = association_rules(frequent_itemsets_ap, metric="confidence", min_threshold=0.8)
rules_ec = association_rules(frequent_itemsets_ap, metric="support", min_threshold=0.2)
rules_fp = association_rules(frequent_itemsets_fp, metric="confidence", min_threshold=0.8)
```
